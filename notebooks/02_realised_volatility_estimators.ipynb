# Realised Volatility Estimators


Compare close-to-close and range-based realised volatility estimators.

```python
import sys
from pathlib import Path

ROOT = Path.cwd()
if ROOT.name == "notebooks":
    ROOT = ROOT.parent
sys.path.insert(0, str(ROOT / "src"))

import duckdb
from volatility_platform.config import DATABASE_PATH
con = duckdb.connect(str(DATABASE_PATH), read_only=True)
rv = con.execute("SELECT * FROM realised_volatility WHERE asset = 'SPY' AND \"window\" = 21").fetchdf()
rv.head()
```

```python
import plotly.express as px
fig = px.line(rv, x='date', y='realised_vol', color='estimator', title='SPY realised volatility estimators')
fig
```
